### Contexte clinique

Le **cancer de la peau** est l'un des cancers les plus frequents au monde. Le **melanome** en represente une minorite mais cause la **majorite des deces** par sa capacite metastatique.

| Stade de detection | Survie a 5 ans |
|---|---|
| **Detecte tot** | **> 95 %** |
| **Detecte tard** | **< 30 %** |

Le diagnostic repose sur la **dermatoscopie** (grossissement a lumiere polarisee). La performance humaine oscille entre **60 % et 90 %** selon l'expertise. Depuis Esteva et al. (*Nature*, 2017), le Deep Learning a egale les **dermatologues experts**.

```python
import os
import random
import shutil
from pathlib import Path

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from PIL import Image

import tensorflow as tf
from sklearn.model_selection import train_test_split

SEED = 42
random.seed(SEED)
np.random.seed(SEED)
tf.random.set_seed(SEED)

IMG_SIZE = 224
BATCH_SIZE = 32
NUM_CLASSES = 7
CLASS_NAMES = ['akiec', 'bcc', 'bkl', 'df', 'mel', 'nv', 'vasc']

sns.set_style('whitegrid')
plt.rcParams['figure.dpi'] = 100

print('TensorFlow version :', tf.__version__)
print('GPU disponible    :', len(tf.config.list_physical_devices('GPU')) > 0)
```

<div style="background: #fff8e1; border-left: 4px solid #ff8f00; padding: 10px 14px; border-radius: 4px; margin: 8px 0; font-size: 13px;">
  <strong style="color: #e65100;">Si <code>kagglehub</code> n'est pas configure :</strong> telecharger manuellement le dataset depuis Kaggle (<code>kmader/skin-cancer-mnist-ham10000</code>) et placer son contenu dans <code>./data/ham10000/</code>.

```python
DATA_DIR = Path('./data/ham10000')

if not DATA_DIR.exists():
    try:
        import kagglehub
        cache_path = Path(kagglehub.dataset_download('kmader/skin-cancer-mnist-ham10000'))
        DATA_DIR.mkdir(parents=True, exist_ok=True)
        for item in cache_path.iterdir():
            target = DATA_DIR / item.name
            if not target.exists():
                if item.is_dir():
                    shutil.copytree(item, target)
                else:
                    shutil.copy2(item, target)
        print('Dataset telecharge vers :', DATA_DIR.resolve())
    except Exception as e:
        print('Echec kagglehub :', e)
        print('Telechargement manuel requis dans', DATA_DIR.resolve())
else:
    print('Dataset deja present :', DATA_DIR.resolve())

for p in sorted(DATA_DIR.iterdir()) if DATA_DIR.exists() else []:
    print(' -', p.name)
```

```python
# Construction de l'index image_id -> chemin (les images sont reparties sur 2 dossiers)
image_dirs = [p for p in DATA_DIR.glob('HAM10000_images*') if p.is_dir()]
image_paths = {}
for d in image_dirs:
    for img in d.glob('*.jpg'):
        image_paths[img.stem] = str(img)

metadata = pd.read_csv(DATA_DIR / 'HAM10000_metadata.csv')
metadata['path'] = metadata['image_id'].map(image_paths)

missing = metadata['path'].isna().sum()
print(f'Total metadonnees : {len(metadata)}')
print(f'Images resolues   : {len(metadata) - missing}')
print(f'Manquantes        : {missing}')
metadata.head()
```

**Resultat** : 4 images exclues sur 10 015 (0,04 %) &mdash; 2 doublons MD5, 2 conflits d'annotation pHash.

```python
# Bascule vers la metadata clean
CLEAN_DIR = Path('./data/ham10000_clean')
clean_meta_path = CLEAN_DIR / 'HAM10000_metadata_clean.csv'

if not clean_meta_path.exists():
    raise FileNotFoundError(
        f"Clean metadata not found at {clean_meta_path}.\n"
        "Run: python clean_dataset.py  (depuis le dossier delivrables/)"
    )

raw_n = len(metadata)
metadata = pd.read_csv(clean_meta_path)
metadata['path'] = metadata['image_id'].map(image_paths)

print(f'Bascule sur le dataset clean')
print(f'  Brut   : {raw_n} images')
print(f'  Clean  : {len(metadata)} images  (- {raw_n - len(metadata)})')

excluded = pd.read_csv(CLEAN_DIR / 'excluded_images.csv')
print(f'\nRaisons d\'exclusion :')
print(excluded['reason'].value_counts().to_string())
```

```python
# --- Nettoyage leger ---
# 57 ages manquants : tagging -1 (non utilise comme feature)
metadata['age'] = metadata['age'].fillna(-1)

# Encodage des labels
label_to_idx = {c: i for i, c in enumerate(CLASS_NAMES)}
metadata['label'] = metadata['dx'].map(label_to_idx)

# --- Split stratifie par lesion_id (anti data leak) ---
df = metadata.dropna(subset=['path']).reset_index(drop=True)

# Une lesion = une seule classe : on agrege au niveau lesion_id
lesions = df.groupby('lesion_id')['label'].first().reset_index()

train_les, temp_les = train_test_split(
    lesions, test_size=0.30, stratify=lesions['label'], random_state=SEED)
val_les, test_les = train_test_split(
    temp_les, test_size=0.50, stratify=temp_les['label'], random_state=SEED)

# Mapping lesion_id -> images
train_df = df[df['lesion_id'].isin(train_les['lesion_id'])].reset_index(drop=True)
val_df   = df[df['lesion_id'].isin(val_les['lesion_id'])].reset_index(drop=True)
test_df  = df[df['lesion_id'].isin(test_les['lesion_id'])].reset_index(drop=True)

# --- Sanity check : aucun lesion_id ne doit etre dans 2 splits ---
assert set(train_df['lesion_id']).isdisjoint(val_df['lesion_id']),  'LEAK train/val'
assert set(train_df['lesion_id']).isdisjoint(test_df['lesion_id']), 'LEAK train/test'
assert set(val_df['lesion_id']).isdisjoint(test_df['lesion_id']),   'LEAK val/test'
print('Anti-leak : OK (aucun lesion_id partage entre splits)')
print()

print(f'Train : {len(train_df):>5} images  ({train_df["lesion_id"].nunique():>5} lesions)')
print(f'Val   : {len(val_df):>5} images  ({val_df["lesion_id"].nunique():>5} lesions)')
print(f'Test  : {len(test_df):>5} images  ({test_df["lesion_id"].nunique():>5} lesions)')

# Verification de la stratification
split_check = pd.DataFrame({
    'train': train_df['dx'].value_counts(normalize=True),
    'val':   val_df['dx'].value_counts(normalize=True),
    'test':  test_df['dx'].value_counts(normalize=True),
}).round(3)
split_check
```

```python
# Pipeline tf.data
def decode_image(path, label):
    img = tf.io.read_file(path)
    img = tf.image.decode_jpeg(img, channels=3)
    img = tf.image.resize(img, [IMG_SIZE, IMG_SIZE])
    img = tf.cast(img, tf.float32) / 255.0
    return img, label

def make_dataset(df_split, training=False):
    paths = df_split['path'].values
    labels = tf.keras.utils.to_categorical(df_split['label'].values, num_classes=NUM_CLASSES)
    ds = tf.data.Dataset.from_tensor_slices((paths, labels))
    ds = ds.map(decode_image, num_parallel_calls=tf.data.AUTOTUNE)
    if training:
        ds = ds.shuffle(2048, seed=SEED)
    ds = ds.batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)
    return ds

train_ds = make_dataset(train_df, training=True)
val_ds   = make_dataset(val_df,   training=False)
test_ds  = make_dataset(test_df,  training=False)

for x, y in train_ds.take(1):
    print('Batch images :', x.shape, x.dtype, 'min', x.numpy().min(), 'max', x.numpy().max())
    print('Batch labels :', y.shape, y.dtype)
```

### Strategie

- **Architecture** : 3 blocs convolutifs (16 &rarr; 32 &rarr; 64 filtres) + FC + softmax
- **Optimiseur** : Adam, `lr = 1e-3`
- **Loss** : entropie croisee categorielle
- **Class weights** : ponderes inversement aux frequences (compensation du desequilibre)
- **Callbacks** : EarlyStopping (patience 5), ReduceLROnPlateau (patience 3), ModelCheckpoint
- **Epochs max** : 20

Tous les modeles partagent les **helpers communs** definis dans la cellule suivante (class weights, callbacks, evaluation, sauvegarde dans `runs/`).

<div style="background: #fff8e1; border-left: 4px solid #ff8f00; padding: 10px 14px; border-radius: 4px; margin: 12px 0; font-size: 13px;">
  <strong>Runtime attendu</strong> : ~30 s/epoch sur GPU, ~3-5 min/epoch sur CPU. Toutes les sorties sont sauvees dans <code>runs/01_baseline/</code>.
</div>

```python
# Helpers communs aux modeles (cellule reutilisee par les sections 6, 7, 8, 11)
import json
from sklearn.utils.class_weight import compute_class_weight
from sklearn.metrics import (classification_report, confusion_matrix,
                              balanced_accuracy_score, f1_score, accuracy_score)

RUNS = Path('./runs')
RUNS.mkdir(exist_ok=True)

# Class weights pour compenser le desequilibre
_w = compute_class_weight('balanced', classes=np.arange(NUM_CLASSES), y=train_df['label'].values)
CLASS_WEIGHTS = dict(enumerate(_w))


def get_callbacks(run_name, patience=5):
    run_dir = RUNS / run_name
    run_dir.mkdir(parents=True, exist_ok=True)
    return [
        tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=patience,
                                          restore_best_weights=True),
        tf.keras.callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.5,
                                              patience=3, min_lr=1e-6),
        tf.keras.callbacks.ModelCheckpoint(str(run_dir / 'best.keras'),
                                            monitor='val_loss', save_best_only=True),
    ]


def evaluate_on(model, ds, label_df):
    y_true = label_df['label'].values
    probs = model.predict(ds, verbose=0)
    y_pred = np.argmax(probs, axis=1)
    f1_arr = f1_score(y_true, y_pred, average=None, zero_division=0)
    return {
        'accuracy':          float(accuracy_score(y_true, y_pred)),
        'balanced_accuracy': float(balanced_accuracy_score(y_true, y_pred)),
        'f1_macro':          float(f1_score(y_true, y_pred, average='macro', zero_division=0)),
        'f1_per_class':      {CLASS_NAMES[i]: float(f1_arr[i]) for i in range(NUM_CLASSES)},
        'confusion_matrix':  confusion_matrix(y_true, y_pred).tolist(),
    }, y_pred


def save_run(run_name, model, history, metrics):
    run_dir = RUNS / run_name
    run_dir.mkdir(parents=True, exist_ok=True)
    model.save(run_dir / 'model.keras')
    with open(run_dir / 'history.json', 'w') as f:
        json.dump({k: [float(x) for x in v] for k, v in history.history.items()}, f, indent=2)
    with open(run_dir / 'metrics.json', 'w') as f:
        json.dump(metrics, f, indent=2)


def plot_curves(run_name, title=None):
    with open(RUNS / run_name / 'history.json') as f:
        h = json.load(f)
    fig, ax = plt.subplots(1, 2, figsize=(12, 3.5))
    ax[0].plot(h['loss'], label='train'); ax[0].plot(h['val_loss'], label='val')
    ax[0].set_title('Loss'); ax[0].set_xlabel('epoch'); ax[0].legend()
    ax[1].plot(h['accuracy'], label='train'); ax[1].plot(h['val_accuracy'], label='val')
    ax[1].set_title('Accuracy'); ax[1].set_xlabel('epoch'); ax[1].legend()
    if title:
        fig.suptitle(title, fontweight='bold')
    plt.tight_layout(); plt.show()


def plot_confusion(metrics, title):
    cm = np.array(metrics['confusion_matrix'])
    fig, ax = plt.subplots(figsize=(7, 5))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=CLASS_NAMES, yticklabels=CLASS_NAMES, ax=ax, cbar=False)
    ax.set_xlabel('Predicted'); ax.set_ylabel('True'); ax.set_title(title)
    plt.tight_layout(); plt.show()


print('Class weights :')
for i, w in CLASS_WEIGHTS.items():
    print(f'  {CLASS_NAMES[i]:>6} -> {w:.2f}')
print(f'\nRuns dir : {RUNS.resolve()}')
```

```python
# Architecture baseline + entrainement
def build_baseline():
    return tf.keras.Sequential([
        tf.keras.layers.Input(shape=(IMG_SIZE, IMG_SIZE, 3)),
        tf.keras.layers.Conv2D(16, 3, activation='relu', padding='same'),
        tf.keras.layers.MaxPooling2D(2),
        tf.keras.layers.Conv2D(32, 3, activation='relu', padding='same'),
        tf.keras.layers.MaxPooling2D(2),
        tf.keras.layers.Conv2D(64, 3, activation='relu', padding='same'),
        tf.keras.layers.MaxPooling2D(2),
        tf.keras.layers.Flatten(),
        tf.keras.layers.Dense(128, activation='relu'),
        tf.keras.layers.Dense(NUM_CLASSES, activation='softmax'),
    ], name='baseline_cnn')

model_baseline = build_baseline()
model_baseline.compile(optimizer=tf.keras.optimizers.Adam(1e-3),
                        loss='categorical_crossentropy', metrics=['accuracy'])
model_baseline.summary()

history_baseline = model_baseline.fit(
    train_ds, validation_data=val_ds,
    epochs=20, class_weight=CLASS_WEIGHTS,
    callbacks=get_callbacks('01_baseline'), verbose=1,
)

metrics_baseline, _ = evaluate_on(model_baseline, test_ds, test_df)
save_run('01_baseline', model_baseline, history_baseline, metrics_baseline)

print(f"\n=== Test set ===")
print(f"Accuracy          : {metrics_baseline['accuracy']:.3f}")
print(f"Balanced accuracy : {metrics_baseline['balanced_accuracy']:.3f}")
print(f"F1 macro          : {metrics_baseline['f1_macro']:.3f}")
print(f"\nF1 par classe :")
for cls, f1 in metrics_baseline['f1_per_class'].items():
    print(f"  {cls:>6} -> {f1:.3f}")
```

    <td style="padding: 10px 14px; background: #fafbfc;">0.25 entre blocs conv, 0.5 dans la tete dense &mdash; combat le surapprentissage</td>
  </tr>
  <tr>
    <td style="padding: 10px 14px; background: #e8f5e9; border-left: 3px solid #2e7d32;"><strong>+ Data augmentation</strong></td>
    <td style="padding: 10px 14px; background: #fafbfc;">Flip H/V, rotation &plusmn;15&deg;, zoom &plusmn;10 %, jitter de contraste &mdash; augmente la diversite percue, surtout utile pour les classes rares</td>
  </tr>

    <td style="padding: 10px 14px; background: #e8f5e9; border-left: 3px solid #2e7d32;"><strong>+ GlobalAveragePooling</strong></td>
    <td style="padding: 10px 14px; background: #fafbfc;">Remplace Flatten &mdash; reduit les parametres, regularise</td>
  </tr>
</table>

Le reste du protocole reste **identique au baseline** pour permettre une **comparaison honnete** : meme optimiseur, meme class weights, memes callbacks, memes epochs max.

```python
# Couche de data augmentation (sera incluse dans le modele pour profiter du GPU)
data_aug = tf.keras.Sequential([
    tf.keras.layers.RandomFlip('horizontal_and_vertical'),
    tf.keras.layers.RandomRotation(0.08),       # +/- 15 degres environ
    tf.keras.layers.RandomZoom(0.1),
    tf.keras.layers.RandomContrast(0.1),
], name='augmentation')


def conv_block(x, filters):
    x = tf.keras.layers.Conv2D(filters, 3, padding='same', use_bias=False)(x)
    x = tf.keras.layers.BatchNormalization()(x)
    x = tf.keras.layers.Activation('relu')(x)
    x = tf.keras.layers.Conv2D(filters, 3, padding='same', use_bias=False)(x)
    x = tf.keras.layers.BatchNormalization()(x)
    x = tf.keras.layers.Activation('relu')(x)
    x = tf.keras.layers.MaxPooling2D(2)(x)
    x = tf.keras.layers.Dropout(0.25)(x)
    return x


def build_improved():
    inp = tf.keras.layers.Input(shape=(IMG_SIZE, IMG_SIZE, 3))
    x = data_aug(inp)
    x = conv_block(x, 32)
    x = conv_block(x, 64)
    x = conv_block(x, 128)
    x = conv_block(x, 256)
    x = tf.keras.layers.GlobalAveragePooling2D()(x)
    x = tf.keras.layers.Dense(128, activation='relu')(x)
    x = tf.keras.layers.Dropout(0.5)(x)
    out = tf.keras.layers.Dense(NUM_CLASSES, activation='softmax')(x)
    return tf.keras.Model(inp, out, name='improved_cnn')


model_improved = build_improved()
model_improved.compile(optimizer=tf.keras.optimizers.Adam(1e-3),
                        loss='categorical_crossentropy', metrics=['accuracy'])
model_improved.summary()

history_improved = model_improved.fit(
    train_ds, validation_data=val_ds,
    epochs=25, class_weight=CLASS_WEIGHTS,
    callbacks=get_callbacks('02_improved', patience=7), verbose=1,
)

metrics_improved, _ = evaluate_on(model_improved, test_ds, test_df)
save_run('02_improved', model_improved, history_improved, metrics_improved)

print(f"\n=== Test set (improved) ===")
print(f"Accuracy          : {metrics_improved['accuracy']:.3f}  (baseline : {metrics_baseline['accuracy']:.3f})")
print(f"Balanced accuracy : {metrics_improved['balanced_accuracy']:.3f}  (baseline : {metrics_baseline['balanced_accuracy']:.3f})")
print(f"F1 macro          : {metrics_improved['f1_macro']:.3f}  (baseline : {metrics_baseline['f1_macro']:.3f})")
```

### Pourquoi VGG16

- **Architecture eprouvee** sur ImageNet, largement utilisee comme backbone en imagerie medicale (litterature HAM10000 abondante).
- **Profondeur raisonnable** (16 couches) &mdash; entrainable sans GPU haut de gamme.
- **Features generiques** (textures, formes) bien transferables aux lesions cutanees.

ResNet50 serait un autre choix legitime (skip connections, plus profond) mais VGG16 offre un **meilleur ratio interpretabilite / cout** pour la section Grad-CAM qui suit.

### Protocole en 2 phases

**Prepro specifique** : on applique la normalisation ImageNet (`tf.keras.applications.vgg16.preprocess_input`) au lieu de `/255`.

```python
# Pipeline dedie : preprocessing VGG16 (au lieu de /255)
from tensorflow.keras.applications.vgg16 import VGG16, preprocess_input as vgg_preprocess

def decode_image_vgg(path, label):
    img = tf.io.read_file(path)
    img = tf.image.decode_jpeg(img, channels=3)
    img = tf.image.resize(img, [IMG_SIZE, IMG_SIZE])
    img = tf.cast(img, tf.float32)
    img = vgg_preprocess(img)  # centre + scale ImageNet
    return img, label

def make_dataset_vgg(df_split, training=False):
    paths = df_split['path'].values
    labels = tf.keras.utils.to_categorical(df_split['label'].values, num_classes=NUM_CLASSES)
    ds = tf.data.Dataset.from_tensor_slices((paths, labels))
    ds = ds.map(decode_image_vgg, num_parallel_calls=tf.data.AUTOTUNE)
    if training:
        ds = ds.shuffle(2048, seed=SEED)
    return ds.batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)

train_ds_vgg = make_dataset_vgg(train_df, training=True)
val_ds_vgg   = make_dataset_vgg(val_df,   training=False)
test_ds_vgg  = make_dataset_vgg(test_df,  training=False)


def build_vgg16():
    backbone = VGG16(weights='imagenet', include_top=False,
                      input_shape=(IMG_SIZE, IMG_SIZE, 3))
    backbone.trainable = False
    inp = tf.keras.layers.Input(shape=(IMG_SIZE, IMG_SIZE, 3))
    x = data_aug(inp)
    x = backbone(x, training=False)
    x = tf.keras.layers.GlobalAveragePooling2D()(x)
    x = tf.keras.layers.Dense(256, activation='relu')(x)
    x = tf.keras.layers.Dropout(0.5)(x)
    out = tf.keras.layers.Dense(NUM_CLASSES, activation='softmax')(x)
    model = tf.keras.Model(inp, out, name='vgg16_tl')
    return model, backbone


model_vgg, backbone_vgg = build_vgg16()
print(f'Total params      : {model_vgg.count_params():,}')
print(f'Trainable params  : {sum(np.prod(v.shape) for v in model_vgg.trainable_variables):,}')
model_vgg.summary()
```

```python
# Phase 1 : warmup tete uniquement (backbone gele)
model_vgg.compile(optimizer=tf.keras.optimizers.Adam(1e-3),
                   loss='categorical_crossentropy', metrics=['accuracy'])

history_vgg_p1 = model_vgg.fit(
    train_ds_vgg, validation_data=val_ds_vgg,
    epochs=10, class_weight=CLASS_WEIGHTS,
    callbacks=get_callbacks('03_vgg_phase1', patience=5), verbose=1,
)

# Phase 2 : fine-tuning du dernier bloc conv (4 dernieres couches)
backbone_vgg.trainable = True
for layer in backbone_vgg.layers[:-4]:
    layer.trainable = False

print(f'\nTrainable params apres degel : '
      f'{sum(np.prod(v.shape) for v in model_vgg.trainable_variables):,}')

model_vgg.compile(optimizer=tf.keras.optimizers.Adam(1e-5),
                   loss='categorical_crossentropy', metrics=['accuracy'])

history_vgg_p2 = model_vgg.fit(
    train_ds_vgg, validation_data=val_ds_vgg,
    epochs=15, class_weight=CLASS_WEIGHTS,
    callbacks=get_callbacks('03_vgg_finetune', patience=7), verbose=1,
)

# Concatener les historiques pour la viz finale
merged_hist = type('H', (), {'history': {
    k: history_vgg_p1.history[k] + history_vgg_p2.history.get(k, [])
    for k in history_vgg_p1.history
}})()

metrics_vgg, _ = evaluate_on(model_vgg, test_ds_vgg, test_df)
save_run('03_transfer_learning', model_vgg, merged_hist, metrics_vgg)

print(f"\n=== Test set (VGG16 fine-tuned) ===")
print(f"Accuracy          : {metrics_vgg['accuracy']:.3f}  (improved : {metrics_improved['accuracy']:.3f})")
print(f"Balanced accuracy : {metrics_vgg['balanced_accuracy']:.3f}  (improved : {metrics_improved['balanced_accuracy']:.3f})")
print(f"F1 macro          : {metrics_vgg['f1_macro']:.3f}  (improved : {metrics_improved['f1_macro']:.3f})")
```

### Classes ciblees

Le sujet demande d'identifier *"une classe sous-representee, difficile ou d'interet clinique particulier"*. On entraine **un GAN par classe rare** pour generer des donnees synthetiques :

| Classe | Images train | Justification |
|--------|--------------|---------------|
| `df` (dermatofibrome) | ~80 | **La plus rare** &mdash; cible prioritaire |
| `vasc` (lesion vasculaire) | ~100 | Egalement tres rare |

`mel` (melanome) est aussi candidate cliniquement mais a deja ~780 images en train &mdash; un GAN y est moins critique.

### Architecture : DCGAN

- **Generateur** : vecteur latent dim 128 &rarr; Dense + Reshape &rarr; 4 blocs Conv2DTranspose &rarr; image RGB 64x64
- **Discriminateur** : 4 blocs Conv2D + LeakyReLU + Dropout &rarr; Dense (sigmoid)
- **Loss** : binary crossentropy (D et G)
- **Optimiseur** : Adam (lr = 2e-4, beta_1 = 0.5) &mdash; recommandation paper DCGAN

### Pourquoi 64x64 et pas 224x224

Generer du 224x224 avec un GAN simple sur 80-100 images est **voue a l'echec** (mode collapse, instabilites). On entraine en **64x64**, on upscale a 224 pour l'enrichissement (au moment de l'injection dans le classifieur).

<div style="background: #fff8e1; border-left: 4px solid #ff8f00; padding: 12px 16px; border-radius: 4px; margin: 16px 0; font-size: 13px;">
  <strong>Runtime attendu</strong> : ~5-8 min par classe sur GPU, ~30-45 min sur CPU. Sorties dans <code>runs/04_gan_{classe}/</code>.
</div>

```python
# Architecture DCGAN 64x64 RGB
GAN_IMG_SIZE = 64
LATENT_DIM = 128
GAN_TARGET_CLASSES = ['df', 'vasc']


def build_generator():
    return tf.keras.Sequential([
        tf.keras.layers.Input(shape=(LATENT_DIM,)),
        tf.keras.layers.Dense(8 * 8 * 256, use_bias=False),
        tf.keras.layers.Reshape((8, 8, 256)),
        tf.keras.layers.BatchNormalization(),
        tf.keras.layers.LeakyReLU(0.2),
        tf.keras.layers.Conv2DTranspose(256, 4, strides=2, padding='same', use_bias=False),
        tf.keras.layers.BatchNormalization(),
        tf.keras.layers.LeakyReLU(0.2),
        tf.keras.layers.Conv2DTranspose(128, 4, strides=2, padding='same', use_bias=False),
        tf.keras.layers.BatchNormalization(),
        tf.keras.layers.LeakyReLU(0.2),
        tf.keras.layers.Conv2DTranspose(64, 4, strides=2, padding='same', use_bias=False),
        tf.keras.layers.BatchNormalization(),
        tf.keras.layers.LeakyReLU(0.2),
        tf.keras.layers.Conv2D(3, 5, padding='same', activation='tanh'),  # 64x64x3
    ], name='generator')


def build_discriminator():
    return tf.keras.Sequential([
        tf.keras.layers.Input(shape=(GAN_IMG_SIZE, GAN_IMG_SIZE, 3)),
        tf.keras.layers.Conv2D(64, 4, strides=2, padding='same'),
        tf.keras.layers.LeakyReLU(0.2),
        tf.keras.layers.Dropout(0.3),
        tf.keras.layers.Conv2D(128, 4, strides=2, padding='same'),
        tf.keras.layers.LeakyReLU(0.2),
        tf.keras.layers.Dropout(0.3),
        tf.keras.layers.Conv2D(256, 4, strides=2, padding='same'),
        tf.keras.layers.LeakyReLU(0.2),
        tf.keras.layers.Dropout(0.3),
        tf.keras.layers.Flatten(),
        tf.keras.layers.Dense(1, activation='sigmoid'),
    ], name='discriminator')


# Test des shapes
g = build_generator(); d = build_discriminator()
z = tf.random.normal([4, LATENT_DIM])
fake = g(z, training=False)
print('Generator output :', fake.shape)
print('Discriminator output :', d(fake, training=False).shape)
del g, d
```

```python
# Boucle d'entrainement DCGAN (custom training loop)
bce = tf.keras.losses.BinaryCrossentropy(from_logits=False)


def make_gan_dataset(class_name, df_split=train_df, n_repeat=8):
    """Charge les images d'une classe, les met en [-1, 1] et repete pour avoir + de batches."""
    sub = df_split[df_split['dx'] == class_name]
    paths = sub['path'].values
    def load(path):
        img = tf.io.read_file(path)
        img = tf.image.decode_jpeg(img, channels=3)
        img = tf.image.resize(img, [GAN_IMG_SIZE, GAN_IMG_SIZE])
        img = (tf.cast(img, tf.float32) / 127.5) - 1.0  # [-1, 1] pour tanh
        return img
    ds = tf.data.Dataset.from_tensor_slices(paths)
    ds = ds.map(load, num_parallel_calls=tf.data.AUTOTUNE).cache()
    ds = ds.repeat(n_repeat).shuffle(512, seed=SEED).batch(32, drop_remainder=True)
    return ds, len(paths)


@tf.function
def train_step(real, gen, disc, g_opt, d_opt):
    bs = tf.shape(real)[0]
    z = tf.random.normal([bs, LATENT_DIM])
    with tf.GradientTape() as gt, tf.GradientTape() as dt:
        fake = gen(z, training=True)
        d_real = disc(real, training=True)
        d_fake = disc(fake, training=True)
        # Label smoothing leger (0.9 au lieu de 1.0)
        d_loss = bce(tf.ones_like(d_real) * 0.9, d_real) + bce(tf.zeros_like(d_fake), d_fake)
        g_loss = bce(tf.ones_like(d_fake), d_fake)
    d_opt.apply_gradients(zip(dt.gradient(d_loss, disc.trainable_variables), disc.trainable_variables))
    g_opt.apply_gradients(zip(gt.gradient(g_loss, gen.trainable_variables), gen.trainable_variables))
    return d_loss, g_loss


def train_gan(class_name, epochs=100):
    print(f'\n=== GAN pour classe `{class_name}` ===')
    ds, n_real = make_gan_dataset(class_name)
    print(f'  Images reelles : {n_real}')

    gen = build_generator()
    disc = build_discriminator()
    g_opt = tf.keras.optimizers.Adam(2e-4, beta_1=0.5)
    d_opt = tf.keras.optimizers.Adam(2e-4, beta_1=0.5)

    losses = {'d': [], 'g': []}
    for ep in range(epochs):
        d_losses, g_losses = [], []
        for real in ds:
            d_l, g_l = train_step(real, gen, disc, g_opt, d_opt)
            d_losses.append(float(d_l)); g_losses.append(float(g_l))
        losses['d'].append(np.mean(d_losses)); losses['g'].append(np.mean(g_losses))
        if (ep + 1) % 10 == 0:
            print(f'  Epoch {ep+1:>3}/{epochs}  d_loss={losses["d"][-1]:.3f}  g_loss={losses["g"][-1]:.3f}')

    out_dir = RUNS / f'04_gan_{class_name}'
    out_dir.mkdir(parents=True, exist_ok=True)
    gen.save(out_dir / 'generator.keras')
    disc.save(out_dir / 'discriminator.keras')
    with open(out_dir / 'losses.json', 'w') as f:
        json.dump(losses, f, indent=2)

    return gen, losses


generators = {}
gan_losses = {}
for cls in GAN_TARGET_CLASSES:
    g, lh = train_gan(cls, epochs=100)
    generators[cls] = g
    gan_losses[cls] = lh
```

```python
# Visualisation : reels vs generes (grille 2 x 8 par classe)
def show_real_vs_fake(class_name, gen, n=8):
    real_paths = train_df[train_df['dx'] == class_name]['path'].sample(n, random_state=SEED).values
    z = tf.random.normal([n, LATENT_DIM])
    fake = gen(z, training=False).numpy()
    fake = ((fake + 1) * 127.5).clip(0, 255).astype(np.uint8)

    fig, axes = plt.subplots(2, n, figsize=(n * 1.6, 4))
    for i, p in enumerate(real_paths):
        img = np.array(Image.open(p).resize((GAN_IMG_SIZE, GAN_IMG_SIZE)))
        axes[0, i].imshow(img); axes[0, i].axis('off')
        axes[1, i].imshow(fake[i]); axes[1, i].axis('off')
    axes[0, 0].set_ylabel('REELLES', fontsize=10, rotation=90, labelpad=12, fontweight='bold')
    axes[1, 0].set_ylabel('GENEREES', fontsize=10, rotation=90, labelpad=12, fontweight='bold', color='#ad1457')
    fig.suptitle(f'GAN classe `{class_name}` - reelles vs generees', fontweight='bold')
    plt.tight_layout()
    plt.savefig(RUNS / f'04_gan_{class_name}' / 'samples.png', dpi=120, bbox_inches='tight')
    plt.show()


for cls in GAN_TARGET_CLASSES:
    show_real_vs_fake(cls, generators[cls])

# Courbes de loss
fig, axes = plt.subplots(1, len(GAN_TARGET_CLASSES), figsize=(6 * len(GAN_TARGET_CLASSES), 3.5))
for ax, cls in zip(np.atleast_1d(axes), GAN_TARGET_CLASSES):
    ax.plot(gan_losses[cls]['d'], label='D loss')
    ax.plot(gan_losses[cls]['g'], label='G loss')
    ax.set_title(f'GAN `{cls}` - loss curves')
    ax.set_xlabel('epoch'); ax.legend()
plt.tight_layout(); plt.show()
```

### Strategie d'injection

- **N = 500** images synthetiques generees par classe rare (`df`, `vasc`) &mdash; on les amene au niveau des classes intermediaires.
- Les images synthetiques sont **upscalees a 224x224** au moment du chargement (bicubique).
- Elles sont **ajoutees au train uniquement**. Val et test restent strictement reels.
- Sauvegarde sur disque dans `data/ham10000_synth/{class}/` pour traceabilite.

<div style="background: #ffebee; border-left: 4px solid #c62828; padding: 12px 16px; border-radius: 4px; margin: 16px 0; font-size: 13px;">
  <strong style="color: #c62828;">Garantie anti-leak</strong> &mdash; <code>val_df</code> et <code>test_df</code> sont strictement composes d'images reelles avec un <code>lesion_id</code> different de ceux du train. Aucune image synthetique ne contamine l'evaluation.
</div>

```python
# Generation et sauvegarde des images synthetiques sur disque
N_SYNTH_PER_CLASS = 500
SYNTH_DIR = Path('./data/ham10000_synth')
SYNTH_DIR.mkdir(parents=True, exist_ok=True)


def generate_and_save(class_name, gen, n=N_SYNTH_PER_CLASS):
    out = SYNTH_DIR / class_name
    out.mkdir(parents=True, exist_ok=True)
    # Generer par batches pour limiter la memoire
    BS = 64
    saved = []
    for batch_start in range(0, n, BS):
        bs = min(BS, n - batch_start)
        z = tf.random.normal([bs, LATENT_DIM])
        fake = gen(z, training=False).numpy()
        fake = ((fake + 1) * 127.5).clip(0, 255).astype(np.uint8)
        for i in range(bs):
            img_id = f'SYNTH_{class_name}_{batch_start + i:04d}'
            path = out / f'{img_id}.jpg'
            Image.fromarray(fake[i]).save(path, quality=95)
            saved.append((img_id, str(path)))
    return saved


# Genere + construit le DataFrame d'enrichissement
synth_records = []
for cls in GAN_TARGET_CLASSES:
    saved = generate_and_save(cls, generators[cls])
    print(f'  {cls:>5} -> {len(saved)} images generees vers {SYNTH_DIR / cls}')
    for img_id, path in saved:
        synth_records.append({
            'image_id': img_id,
            'lesion_id': img_id,           # ID unique fictif
            'dx': cls,
            'dx_type': 'synth',
            'age': -1, 'sex': 'synth', 'localization': 'synth',
            'dataset': 'synth_gan',
            'path': path,
            'label': CLASS_NAMES.index(cls),
        })

synth_df = pd.DataFrame(synth_records)
train_df_enriched = pd.concat([train_df, synth_df], ignore_index=True)

print(f'\nTrain set :')
print(f'  Reel       : {len(train_df):>5}')
print(f'  Synthetique: {len(synth_df):>5}')
print(f'  Enrichi    : {len(train_df_enriched):>5}')
print(f'\nDistribution apres enrichissement :')
print(train_df_enriched['dx'].value_counts().reindex(CLASS_NAMES).to_string())
```

### Pourquoi Grad-CAM

Grad-CAM (Selvaraju et al., 2017) produit une **carte thermique** indiquant les regions de l'image qui ont **le plus contribue a la prediction**. C'est l'outil standard pour ouvrir la boite noire d'un CNN.

### Analyse cible : trois cas

<table style="width:100%; border-collapse: collapse; margin: 12px 0;">
  <tr>
    <td style="padding: 10px 14px; background: #e8f5e9; border-left: 3px solid #2e7d32; width: 25%;"><strong>1. Cas reussis</strong></td>
    <td style="padding: 10px 14px; background: #fafbfc;">Predictions correctes avec haute confiance. <strong>Verifier</strong> que le modele regarde la lesion (pas l'arriere-plan).</td>
  </tr>
  <tr>
    <td style="padding: 10px 14px; background: #ffebee; border-left: 3px solid #c62828;"><strong>2. Cas errones</strong></td>
    <td style="padding: 10px 14px; background: #fafbfc;">Predictions fausses. <strong>Comprendre pourquoi</strong> &mdash; le modele s'est concentre sur quoi ?</td>
  </tr>
  <tr>
    <td style="padding: 10px 14px; background: #fff3e0; border-left: 3px solid #e65100;"><strong>3. Detection de biais</strong></td>
    <td style="padding: 10px 14px; background: #fafbfc;">Cas ou Grad-CAM s'allume sur des <strong>artefacts</strong> (regle graduee, encre, poils). Demonstration empirique des biais documentes.</td>

On utilise le **meilleur modele** (VGG16 enrichi par GAN), couche cible = `block5_conv3` (la derniere conv du backbone).

```python
# Implementation Grad-CAM (compatible Keras 3 + modeles imbriques)
import cv2

def get_grad_cam_model(model, last_conv_layer_name='block5_conv3'):
    """Construit un modele qui sort (conv_features, predictions) en flat,
    meme si le modele original contient des sous-modeles (VGG16 imbrique)."""
    # 1) Localise le sous-modele qui contient la couche cible
    target_submodel = None
    submodel_idx = None
    for i, l in enumerate(model.layers):
        if hasattr(l, 'get_layer'):
            try:
                l.get_layer(last_conv_layer_name)
                target_submodel = l
                submodel_idx = i
                break
            except (ValueError, KeyError):
                continue
    if target_submodel is None:
        raise ValueError(f"Layer {last_conv_layer_name} not found in any nested submodel")

    last_conv = target_submodel.get_layer(last_conv_layer_name)

    # 2) Sous-modele qui expose (conv_features, sortie_du_backbone)
    sub_grad = tf.keras.Model(
        inputs=target_submodel.input,
        outputs=[last_conv.output, target_submodel.output],
    )

    # 3) Reconstruction flat de l'outer model
    inp = tf.keras.Input(shape=model.input_shape[1:])
    x = inp
    for l in model.layers[:submodel_idx]:
        if not isinstance(l, tf.keras.layers.InputLayer):
            x = l(x, training=False)
    conv_out, backbone_out = sub_grad(x)
    h = backbone_out
    for l in model.layers[submodel_idx + 1:]:
        if not isinstance(l, tf.keras.layers.InputLayer):
            h = l(h, training=False) if 'training' in l.call.__code__.co_varnames else l(h)
    return tf.keras.Model(inputs=inp, outputs=[conv_out, h])


def make_gradcam_heatmap(img_array, grad_model, pred_index=None):
    with tf.GradientTape() as tape:
        conv_out, preds = grad_model(img_array)
        if pred_index is None:
            pred_index = tf.argmax(preds[0])
        class_channel = preds[:, pred_index]
    grads = tape.gradient(class_channel, conv_out)
    pooled_grads = tf.reduce_mean(grads, axis=(0, 1, 2))
    conv_out = conv_out[0]
    heatmap = conv_out @ pooled_grads[..., tf.newaxis]
    heatmap = tf.squeeze(heatmap).numpy()
    heatmap = np.maximum(heatmap, 0) / (heatmap.max() + 1e-9)
    return heatmap, int(pred_index), float(preds[0, pred_index])


def overlay_heatmap(img_rgb, heatmap, alpha=0.4):
    heatmap = cv2.resize(heatmap, (img_rgb.shape[1], img_rgb.shape[0]))
    heatmap = np.uint8(255 * heatmap)
    colored = cv2.applyColorMap(heatmap, cv2.COLORMAP_JET)
    colored = cv2.cvtColor(colored, cv2.COLOR_BGR2RGB)
    return np.uint8(img_rgb * (1 - alpha) + colored * alpha)


# On utilise le modele VGG16 enrichi (meilleur)
gradcam_model = get_grad_cam_model(model_vgg_enr)


def show_gradcam_grid(rows, title):
    """rows = liste de tuples (image_path, true_label, comment)"""
    fig, axes = plt.subplots(len(rows), 3, figsize=(11, 3.2 * len(rows)))
    if len(rows) == 1:
        axes = axes[np.newaxis, :]
    for i, (path, true_label, comment) in enumerate(rows):
        pil = Image.open(path).resize((IMG_SIZE, IMG_SIZE))
        img_rgb = np.array(pil)
        arr = vgg_preprocess(np.expand_dims(img_rgb.astype(np.float32), 0))
        heatmap, pred_idx, conf = make_gradcam_heatmap(arr, gradcam_model)
        overlay = overlay_heatmap(img_rgb, heatmap)

        axes[i, 0].imshow(img_rgb); axes[i, 0].axis('off')
        axes[i, 0].set_title(f'Vraie: {true_label}', fontsize=10)
        axes[i, 1].imshow(heatmap, cmap='jet'); axes[i, 1].axis('off')
        axes[i, 1].set_title('Grad-CAM', fontsize=10)
        axes[i, 2].imshow(overlay); axes[i, 2].axis('off')
        color = '#2e7d32' if CLASS_NAMES[pred_idx] == true_label else '#c62828'
        axes[i, 2].set_title(f'Pred: {CLASS_NAMES[pred_idx]} ({conf:.2f})\n{comment}',
                              fontsize=10, color=color)
    fig.suptitle(title, fontweight='bold', y=1.005)
    plt.tight_layout(); plt.show()


# 1) Cas reussis : un par classe, predit correctement avec haute confiance
print('Calcul des predictions sur le test set...')
probs_test = model_vgg_enr.predict(test_ds_vgg, verbose=0)
test_pred = np.argmax(probs_test, axis=1)
test_conf = probs_test.max(axis=1)

successes = []
for cls in CLASS_NAMES:
    cls_idx = CLASS_NAMES.index(cls)
    mask = (test_df['label'].values == cls_idx) & (test_pred == cls_idx)
    if mask.sum() > 0:
        idxs = np.where(mask)[0]
        best = idxs[np.argmax(test_conf[idxs])]
        successes.append((test_df.iloc[best]['path'], cls, 'Bien classe'))
show_gradcam_grid(successes, '1. Cas reussis - 1 par classe, plus haute confiance')
```

```python
# 2) Cas errones les plus confiants (le modele se trompe avec haute conviction)
mask_err = (test_pred != test_df['label'].values)
err_idxs = np.where(mask_err)[0]
err_top = err_idxs[np.argsort(-test_conf[err_idxs])][:6]

errors = []
for i in err_top:
    row = test_df.iloc[i]
    pred_cls = CLASS_NAMES[test_pred[i]]
    errors.append((row['path'], row['dx'], f'confondu avec {pred_cls}'))
show_gradcam_grid(errors, '2. Cas errones - les 6 erreurs les plus confiantes')
```

```python
# 3) Detection de biais : images du test contenant probablement des artefacts
# Heuristique : forte composante en lignes droites (regles) ou bord sombre
def has_ruler_like(path, threshold=80):
    """Detecte lignes droites via Hough transform - proxy pour regles graduees."""
    img = cv2.imread(str(path), 0)
    img = cv2.resize(img, (224, 224))
    edges = cv2.Canny(img, 50, 150)
    lines = cv2.HoughLinesP(edges, 1, np.pi/180, threshold=threshold,
                            minLineLength=50, maxLineGap=10)
    return len(lines) if lines is not None else 0


print('Recherche d\'images avec potentiels artefacts de regles dans le test set...')
test_df['n_lines'] = test_df['path'].apply(has_ruler_like)
suspect_candidates = test_df.sort_values('n_lines', ascending=False).head(6)

bias_cases = [(row['path'], row['dx'], f'{int(row["n_lines"])} lignes detectees')
              for _, row in suspect_candidates.iterrows()]
show_gradcam_grid(bias_cases, '3. Suspects de biais - images avec le plus d\'arretes droites')
```

### Ameliorations possibles

- **GAN plus avance** : StyleGAN2-ADA est concu pour les datasets de petite taille (~few-shot generation).
- **cGAN conditionne** : un seul GAN qui genere les 7 classes au lieu de N GANs separes.
- **Diffusion models** : SOTA pour la generation medicale en 2024+, plus stable que les GAN.
- **Generation a 224x224** avec progressive growing au lieu de l'upscale.
- **Ensemble de modeles** : moyenner TL et TL+GAN au lieu de choisir l'un ou l'autre.
- **Stratification par tonalite de peau** (Fitzpatrick scale) pour quantifier le biais demographique.

### Principaux enseignements

1. **L'accuracy globale est trompeuse** sur dataset desequilibre &mdash; la **balanced accuracy** et le **F1 par classe** sont les vraies metriques d'interet clinique.
2. Le **Transfer Learning** apporte un gain massif vs from scratch, meme avec un dataset de 10k images.
3. Le **GAN n'est utile que conditionnellement** &mdash; son apport sur les classes rares depend de la qualite des images generees et de la diversite preservee. Une demarche aveugle "plus de donnees = mieux" ne marche pas.
4. **Grad-CAM revele des biais d'acquisition** &mdash; un modele performant en chiffres peut etre cliniquement non fiable si ses decisions reposent sur des artefacts.
5. La **rigueur methodologique** (split par lesion_id, separation stricte train/val/test, audit qualite) est aussi importante que le choix d'architecture.

### Perspectives

- **Cross-validation stratifiee par lesion_id** pour estimer la variance des metriques
- **Datasets externes** (ISIC 2019, BCN20000) pour mesurer la generalisation
- **Diffusion models** au lieu de DCGAN
- **Evaluation par tonalite de peau** (Fitzpatrick) pour quantifier l'equite demographique
- **Etude de calibration** (reliability diagrams, ECE) &mdash; un modele clinique doit produire des probabilites bien calibrees
